```python
# Replacing missing values with average values based on State and Year
for x in dataNew:
    if dataNew[x].dtypes!='object' and dataNew[x].isnull().sum()>0:
        #Total[x]=Total.groupby("").transform(lambda x: x.fillna(x.mean()))
        #print(dataNew2.loc[dataNew[x]=="BOYS 7-12 YEARS"])
        #print(x)
        # Replace missing data with features mean
        dataNew[x]=dataNew[x].fillna(dataNew.groupby(["STATE/UT","YEAR"])[x].transform('mean'))
```

```python
# Handling missing data in Categorical features

for i in CategoricalData:
    if dataNew[i].isnull().sum()>0:
        #print(i)
        #dataNew[i]=dataNew.groupby(["STATE/UT","YEAR"]).transform(lambda x: x.fillna(x.mean()))
        dataNew[i]=dataNew[i].fillna(dataNew.groupby(["STATE/UT","YEAR"])[x].transform('mean'))
    else:
        dataNew[i]=dataNew[i].fillna(dataNew.groupby("STATE/UT")[x].transform('mean'))
```
